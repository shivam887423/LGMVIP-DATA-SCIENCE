# **Preparing Data for EDA Analysis**

```python
#rename some columns to get better visualization
data.rename(columns={'eventid':'Eventid', 'provstate':'State','iyear':'Year','imonth':'Month','city':'City','region':'Region',
                   'suicide':'Suicide','crit1':'City1','attacktype1':'Attack_type','targtype1':'Target','nkill':'Killed','natlty1':'Nationality','nwound':'Wounded','summary':'Summary',
                   'gname':'Group','targtype1_txt':'Target_type','claimed':'Claimed','weaptype1_txt':'Weapon_type','crit2':'City2','crit3':'City3','guncertain1':'Guntype'},inplace=True)
data
```

```python
data_list=list(data)
print(data_list)
```

```python
# Taking out required Columns for Analysis
data_1=data[['Eventid', 'Year', 'Month','Country', 'Region','State','City','City1', 'City2', 'City3', 'Suicide', 'Attack_type','Target','Nationality','Group','Guntype','Claimed','Weapon_type','Killed','Wounded']]
data_1.head()
```

```python
# Combining Two Columns i.e., 'nkill' and 'nwound' into a new column 'casualities'
data_1['Casualities'] = data_1['Killed'] + data_1['Wounded']
```

```python
data_1.columns
```

# **Checking null values for the new dataset...**

```python
data_1.isnull().sum()
```

```python
#Drop all the null values
# Here i give the new dataset named "Terror_Final" after dropping all the null values

data_1 = data_1.dropna(axis=0)
```

#**DATA VISUALIZATION**

# 1 Number Of Terrorist Activities Held  Each Year ? 

```python
plt.figure(figsize=(15,8))
sns.set_context("poster" , font_scale = 0.7)
sns.countplot(data_1["Year"],palette='RdYlGn_r',)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year' ,fontsize = 30 )
plt.xlabel('Year' , fontsize = 20)
plt.ylabel('Count' , fontsize = 20)
plt.show()
```

#**INFERENCE**
From above Graph:_


*  In the year 2014 most terrorist attacked were held.
*  The year between 1973 and 1997 the terrorist attack were very least.





# 2 What were the casualities Between Each Year ?

```python
yearc=data[['Year','Casualities']].groupby('Year').sum()
yearc.plot(kind='bar',color='green',figsize=(15,6))
plt.title("Casualities")
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Casualities',fontsize=15)
plt.show()
```

# **INFERENCE**
From above graph

*   Most of the casualities were happened in the year 2015.
*   The least casualities wert from year 1971. 



# 3 On Which Type Of Wepon were Mostly Attacked

```python
plt.figure(figsize=(12,6))
sns.countplot(y=data_1['Weapon_type'])
plt.title('Count of attacks with various Weapons ' , fontsize = 30)
plt.xlabel('Counts' , fontsize = 30)
plt.ylabel('Weapon_Type', fontsize = 30 )
plt.show()
```

# **INFERENCE**


*   So the **explosive** were the main weapons which were use for attack.
*   Weapons like **Radiological, Biological** were not use for attacks 



# 4 What are The Numbers Totals Attacks in Each Year ?

```python
plt.subplots(figsize=(15,6))
country_attacks = data_1.Group.value_counts()[:15].reset_index()
country_attacks.columns = ['Group', 'Total Attacks']
sns.barplot(x = country_attacks.Group, y = country_attacks['Total Attacks'], palette='OrRd_r', edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation = 90)
plt.title("Number of Total Attacks in Each Country")
plt.show()
```

# 5  The Most Active Terrorist Organization ?

```python
fig=plt.figure(figsize=(20,10))
sns.barplot(x=data_1['Group'].value_counts()[1:11].values,
            y=data_1['Group'].value_counts()[1:11].index,
            palette='inferno')
plt.yticks(size=15)
plt.xlabel("No. of Attacks ",fontsize=25)
plt.ylabel("Terrorist Organization",fontsize=25)
plt.title("Most Active Terrorist Organization",fontsize=35)
plt.show()
```

## **INFERENCE**


*   Taliban is the most active Terrorist Organisation with 6000+ atacks number .




# 6 Which are The Top Affected Years ?

```python
plt.subplots(figsize=(15,8))
sns.barplot(data_1['Year'].value_counts()[:10].index,
            data_1['Year'].value_counts()[:10].values,
            palette='Blues_d')
plt.title('Top Years Affected' , fontsize = 35)
plt.xlabel('Years' , fontsize = 30)
plt.ylabel('Count' , fontsize = 30)
plt.xticks(rotation= 90 , fontsize = 20)
plt.show()
```

# **INFERENCE**



*   The year 2014 was mostly affected.
*   The years 2008 to 2011 were equally affected.
